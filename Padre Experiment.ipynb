```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score
from pypadre.app import p_app
from pypadre.ds_import import load_csv
from pypadre.core import Experiment
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from pypadre.experimentexecutor import ExperimentExecutor

ds = load_csv('ds2new.csv',targets=['calving difficulty'],name=None, description="imported form csv", source="csvloaded")
print(ds)
```

```python

# Create an experiment with multiple runs
params_pca = {'num_components': [2, 3, 4, 5]}
params_svc = {'C': [0.5, 1.0, 1.5],
'poly_degree': [1, 2, 3]}
params_dict = {'SVC': params_svc, 'pca': params_pca}
workflow = p_app.experiment_creator.create_test_pipeline(['pca', 'SVC'])
#p_app.experiment_creator.clear_experiments('HyperParameter Search')
p_app.experiment_creator.create(name='Experiment for finding optimal parameters on PCA, SVR',
description='Experiment to test the comparison of metrics',
dataset_list=[ds],
workflow=workflow,
params=params_dict
)
experiments_executor = ExperimentExecutor(experiments=p_app.experiment_creator.createExperimentList())
experiments_executor.execute(local_run=True, threads=1)
```

```python
def create_test_pipeline_RF():
    estimators = [('Random Forest', RandomForestClassifier())]
    return Pipeline(estimators)

p_app.experiment_creator.create(name='Random Forest',
description='Simple iPython notebook experioment',
workflow=create_test_pipeline_RF(),
dataset_list=[ds])
p_app.experiment_creator.execute()
```

```python
def create_test_pipeline_DT():
    estimators = [('Decision Tree', DecisionTreeClassifier())]
    return Pipeline(estimators)


p_app.experiment_creator.create(name='Decision Tree experiment',
description='Simple iPython notebook experioment',
workflow=create_test_pipeline_DT(),
dataset_list=[ds])
p_app.experiment_creator.execute()
```
